# Monte Carlo Methods on OpenAI Gym's Blackjack

```python
import blackjack
from mpl_toolkits.axes_grid1 import make_axes_locatable
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Set Training Variables
Set num_episodes and epsilon_decay values.
Calculate maximum and minimum epsilon values.

```python
num_episodes = 100000
epsilon_decay = 8000
max_epsilon = 1.0/((1/epsilon_decay) + 1)
min_epsilon = 1.0/((num_episodes/epsilon_decay) + 1)
print("Max epsilon value: {:0.6f}".format(max_epsilon))
print("Min epsilon value: {:0.6f}".format(min_epsilon))
```

## Plot Epsilon Decay

```python
# plot epsilon decay
x = np.linspace(0,num_episodes,100)
y = 1.0/((x/epsilon_decay) + 1)

plt.xlabel('Episode Number')
plt.ylabel('Epsilon Value')

lines = plt.plot(x,y)
plt.setp(lines, 'color', 'g',)

plt.title('Decay of Epsilon Values Over Time', fontsize=16)

plt.savefig('epsilon_values.png')
plt.show()
```

## Train

```python
policy, Q = blackjack.learn(num_episodes=num_episodes, epsilon_decay=epsilon_decay)
```

## Score

```python
final_average_return = blackjack.score(policy)
print("final average returns: {}".format(final_average_return))
```

## Plot

```python
blackjack.plot_policy(policy, "diag_{}_{}_{}.png".format(num_episodes, epsilon_decay, final_average_return))
```
